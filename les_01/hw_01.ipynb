4. Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)

k1 = 1
k2 = 2

y1 = np.cos(k1 * x)
y2 = np.cos(k2 * x)

plt.plot(x, y1)
plt.plot(x, y2)

plt.title("Графики функций")
plt.xlabel("Переменная x")
plt.ylabel("Функция y(x)")

plt.legend(labels=["k1", "k2"])

plt.show()
```
