1) окружности

```python
x = 4
y = 3  
R = 2   

t = np.linspace(0, 2 * np.pi, 100)

plt.plot( x + R * np.cos(t), y + R * np.sin(t))

plt.axis('scaled')

plt.show()
```

```python
x = 1
y = 2   
a = 6   
b = 4

t = np.linspace(0, 2 * np.pi, 100)

plt.plot( x + a * np.cos(t), y + b * np.sin(t))

plt.axis('scaled')

plt.show()
```

3) гиперболы

```python
x = 1
y = 2   
a = 1   
b = 3

t = np.linspace(-10, 10, 100)
ch_t = (np.exp(t) + np.exp(-t)) / 2
sh_t = (np.exp(t) - np.exp(-t)) / 2

plt.plot(x + a * ch_t, y + b * sh_t)

plt.xlim(-10, 10) 
plt.ylim(-10, 10)

plt.show()
```

##### 1. Задание

Нарисуйте график функции: y(x) = k∙cos(x – a) + b для некоторых (2-3 различных) значений параметров k, a, b

```python
x = np.linspace(0, 10, 100)

k1 = 1
k2 = 2
a1 = 3
a2 = 5
b1 = 2
b2 = 10

y1 = k1 * np.cos(x - a1) + b1
y2 = k2 * np.cos(x - a2) + b2

plt.plot(x, y1)
plt.plot(x, y2)

plt.title("Графики функций")
plt.xlabel("Переменная x")
plt.ylabel("Функция y(x)")

plt.legend(labels=["y1", "y2"])

plt.show()
```

1) Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polar_to_decart(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x,y
```

```python
x = np.linspace(0, 15, 100)
y = 8 * x
plt.polar(x, y)

plt.show()
```

```python
r = np.linspace(0, 10, 100)
phi = np.linspace(0, np.pi * 2, 100)

x, y = polar_to_decart(r, phi)

plt.plot(x, 8 * x)

plt.xlim(-5, 5) 
plt.ylim(-20, 20)
plt.grid(True)

plt.show()
```

1) Решите систему уравнений:

y = x^2 – 1

exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-2, 3, 201)

plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) - 1)/x + 1)

plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3,8) 
plt.grid(True)

plt.show()
```

```python
def equations(p):
    x, y = p
    return (x**2 - 1 - y, (np.exp(x) - 1)/x +1 - y)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (-1, 2))
              
print (x1, y1)
print (x2, y2)
```
