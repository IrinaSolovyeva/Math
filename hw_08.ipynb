## Задача 1

Используя данные о хоккеистах (скачать можно здесь), проверьте, является ли среди хоккеистов из Финляндии, Норвегии и Дании значимым отличие: а) роста, б) BMI

```python
import numpy as np
import pandas as pd
from warnings import filterwarnings
filterwarnings("ignore")
from scipy import stats
```

```python
df = pd.read_csv('hockey_players.csv', encoding='ISO-8859-1')

df
```

```python
df.info()
```

```python
df.describe().transpose()
```

```python
df.select_dtypes(include='object').describe().transpose()
```

```python
df.drop_duplicates(['name'], inplace=True)
```

```python
df.select_dtypes(include='object').describe().transpose()
```

```python
FIN_height = df.loc[(df['country'] == 'FIN'), 'height']
NOR_height = df.loc[(df['country'] == 'NOR'), 'height']
DEN_height = df.loc[(df['country'] == 'DEN'), 'height']
```

```python
stats.f_oneway(FIN_height, NOR_height, DEN_height)
```

### Ответ:
P-значение меньше уровня значимости, следовательно, отличие роста между выборками спортсменов из Финляндии, Норвегии и Дании признается значимым

```python
FIN_bmi = df.loc[(df['country'] == 'FIN'), 'bmi']
NOR_bmi = df.loc[(df['country'] == 'NOR'), 'bmi']
DEN_bmi = df.loc[(df['country'] == 'DEN'), 'bmi']
```

```python
stats.f_oneway(FIN_bmi, NOR_bmi, DEN_bmi)
```

### Ответ:
P-значение меньше уровня значимости, следовательно, отличие BMI между выборками спортсменов из Финляндии, Норвегии и Дании признается значимым
